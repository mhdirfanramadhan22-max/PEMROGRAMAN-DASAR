```python
# Program klasifikasi beban listrik

# Input data
kapasitas_maks = float(input("Masukkan kapasitas maksimum (Watt): "))
beban_terpakai = float(input("Masukkan beban terpakai (Watt): "))

# Hitung persentase pemakaian
persentase = (beban_terpakai / kapasitas_maks) * 100

# Klasifikasi berdasarkan persentase
if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"

# Tampilkan hasil
print(f"\nPersentase pemakaian: {persentase:.2f}%")
print(f"Status beban: {status}")
```

```python
# PROGRAM MONITORING PEMBANGKIT LISTRIK
# Mengimplementasikan percabangan, perulangan, dan fungsi

# ===== FUNGSI UTAMA =====
def hitung_efisiensi(beban_terpakai, kapasitas_maksimum):
    """Fungsi untuk menghitung efisiensi pembangkit"""
    if kapasitas_maksimum <= 0:
        return 0
    return (beban_terpakai / kapasitas_maksimum) * 100


def klasifikasikan_beban(persentase):
    """Fungsi untuk mengklasifikasikan beban berdasarkan persentase"""
    if persentase <= 60:
        return "Beban Ringan"
    elif persentase <= 85:
        return "Beban Sedang"
    else:
        return "Beban Berat"


def cek_status_suhu(suhu):
    """Fungsi untuk mengecek status suhu"""
    if suhu <= 90:
        return "Normal"
    elif suhu < 100:
        return "Warning"
    else:
        return "SHUTDOWN"


# ===== DATA SENSOR =====
data_sensor = {
    'suhu_boiler': [92.5, 88.7, 95.1, 102.3, 97.8],
    'vibrasi_mesin': [1.2, 0.9, 2.5, 3.1, 1.8],
    'beban_unit': [45, 75, 90, 60, 85]  # dalam MW
}

# ===== PROSES ANALISIS =====
kapasitas_maksimum = 100  # Misalnya kapasitas maksimum 100 MW

print("=== HASIL MONITORING PEMBANGKIT LISTRIK ===\n")

for i in range(len(data_sensor['suhu_boiler'])):
    suhu = data_sensor['suhu_boiler'][i]
    beban = data_sensor['beban_unit'][i]

    efisiensi = hitung_efisiensi(beban, kapasitas_maksimum)
    status_beban = klasifikasikan_beban(efisiensi)
    status_suhu = cek_status_suhu(suhu)

    print(f"Unit {i+1}:")
    print(f"  Suhu Boiler     : {suhu}°C ({status_suhu})")
    print(f"  Beban Terpakai  : {beban} MW ({status_beban})")
    print(f"  Efisiensi       : {efisiensi:.2f}%\n")
```
